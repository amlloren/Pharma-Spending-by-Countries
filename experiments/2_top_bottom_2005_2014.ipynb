#### Supporting Python Code:

```python
# Filter the DataFrame to include only the years 2005 to 2014
df_tfiltered = df[(df['TIME'] >= 2005) & (df['TIME'] <= 2014)]

# Exclude the dropped countries
df_tfiltered = df_tfiltered[~df_tfiltered['LOCATION'].isin(['GBR', 'NZL', 'TUR', 'RUS'])]

# Identify the Top 10 highest spending countries (Per Capita)
top_10_per_cap = df_tfiltered.groupby('LOCATION')['USD_CAP'].mean().sort_values(ascending=False).head(10)

# Display the results
print("Top 10 Highest Spending Countries:")
print(top_10_per_cap)
```

```python
# Filter the DataFrame to include only the years 2005 to 2014
df_bfiltered = df[(df['TIME'] >= 2005) & (df['TIME'] <= 2014)]

# Exclude the dropped countries
df_bfiltered = df_bfiltered[~df_bfiltered['LOCATION'].isin(['GBR', 'NZL', 'TUR', 'RUS'])]

# Identify the Bottom 10 highest spending countries (Per Capita)
bottom_10_per_cap = df_bfiltered.groupby('LOCATION')['USD_CAP'].mean().sort_values(ascending=True).head(10)

# Display the results
print("Bottom 10 Lowest Spending Countries (Excluding dropped countries):")
print(bottom_10_per_cap)
```

## Data Visualization using Python

```python
# Filter the DataFrame to include only the years 2005 to 2014
df_tfiltered = df[(df['TIME'] >= 2005) & (df['TIME'] <= 2014)]

# Exclude the dropped countries
df_tfiltered = df_tfiltered[~df_tfiltered['LOCATION'].isin(['GBR', 'NZL', 'TUR', 'RUS'])]

# Identify the Top 10 highest spending countries (Per Capita)
top_10_per_cap = df_tfiltered.groupby('LOCATION')['USD_CAP'].mean().sort_values(ascending=False).head(10)

# Plot the bar chart for top 10 spenders
plt.figure(figsize=(10, 6))
plt.bar(top_10_per_cap.index, top_10_per_cap.values, color='skyblue')
plt.title('Top 10 Countries by Health Expenditure Per Capita (2005-2014)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Spending Per Capita ($ USD)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame to include only the years 2005 to 2014
df_bfiltered = df[(df['TIME'] >= 2005) & (df['TIME'] <= 2014)]

# Exclude the dropped countries
df_bfiltered = df_bfiltered[~df_bfiltered['LOCATION'].isin(['GBR', 'NZL', 'TUR', 'RUS'])]

# Identify the Bottom 10 highest spending countries (Per Capita)
bottom_10_per_cap = df_bfiltered.groupby('LOCATION')['USD_CAP'].mean().sort_values(ascending=True).head(10)

# Plot the bar chart for top 10 spenders
plt.figure(figsize=(10, 6))
plt.bar(bottom_10_per_cap.index, bottom_10_per_cap.values, color='skyblue')
plt.title('Bottom 10 Countries by Health Expenditure Per Capita (2005-2014)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Spending Per Capita ($ USD)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```
